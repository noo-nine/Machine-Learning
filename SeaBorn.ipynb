# **Seaborn with Pandas**

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
diamonds.dropna(inplace = True)
diamonds.shape
```

```python
sns.relplot(data = diamonds,x = 'cut', y = 'price', col = 'color', hue = 'clarity')
```

# **KDEPlot(Kernel Density Estimation)**

```python
# we use gaussian mostly
# 1D or univariate plot
diamonds.head()
```

```python
sns.kdeplot(data = diamonds, x = 'price')
```

```python
sns.kdeplot(data = diamonds, x = 'price', fill = True)
```

```python
sns.kdeplot(data = diamonds, x = 'price', bw_adjust = 5)
# more smooth -- oversmoothed
```

```python
sns.kdeplot(data = diamonds, x = 'price', bw_adjust = 0.5)
# not smooth -- undersmooth
```

```python
sns.kdeplot(data = diamonds, x = 'price', cumulative = True)
```

```python
# bivariate kdeplot
# 2D
sns.kdeplot(data = diamonds, x = diamonds.price, y = diamonds.depth, n_levels = 20)
```

```python
sns.kdeplot(data = diamonds, x = diamonds.price, y = diamonds.depth, cmap = 'Reds', fill = True)
```

# **BOX PLOT**

```python
# box s formed by taking median of points on a line, thhen medians of those 2 parts
# again and joining the 3 lines --- 25th percentile, median, 75th percentile
# other part is 'whiskers'
# starts by measuring inner quartile range(IQR) -> distance between 25th and 75th precentiles
# stretch the wiskers in either directions of 1.5.(IQR)
# data points outside the whisker range --> "outliers"
```

```python
sns.set_style('whitegrid')
```

```python
sns.boxplot(data = diamonds, x = 'price')
```

```python
sns.boxplot( data = diamonds, x = 'cut', y = 'carat')
```

# **Hue**

```python
sns.boxplot( data = diamonds, x = 'cut', y = 'carat', hue = 'color')
```

```python
# hue can be used efficiently by using 'true; and 'false' values
# we can create our own columns by using pandas if true and false columns are not already present
```

# **Styling**

```python
# bars can be converted from vertical to horizontal and vice versa by changing their axes
sns.boxplot( data = diamonds, x = 'carat', y = 'cut')
```

```python
# to order the bars the way we like
sns.boxplot( data = diamonds, x = 'carat', y = 'cut', order = ['Fair', 'Good', 'Very Good','Ideal', 'Premium'])
```

```python
# hue order can also be changed in the same way
```
