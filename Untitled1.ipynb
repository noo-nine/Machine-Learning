# DOWNSAMPLING THE DATA

```python
# since SVMs are good with small datasets
# and there are 2 categories of 'default' -- defaulting with credit card, not defaulting
# we will take 1000 samples each from a category
```

```python
credit_default = credit_cleaned[ credit_cleaned['default'] == 1]
credit_not_default = credit_cleaned[credit_cleaned['default'] == 0]
```

```python
#we split the credit card data into 2
# we will downsample each with 1000 rows
# and merge them to create a model on it
```

```python
from sklearn.utils import resample

credit_default_downsampled = resample( credit_default, replace = False, n_samples = 1000, random_state = 21)
credit_not_default_downsampled = resample( credit_not_default, replace = False, n_samples = 1000, random_state = 21)
```

```python
len(credit_default_downsampled)
```

```python
len(credit_not_default_downsampled)
```

```python
# concatinating both the downsampled data
credit_downsampled = pd.concat([credit_default_downsampled, credit_not_default_downsampled])
len(credit_downsampled)
```

```python
X = credit_downsampled.drop('default', axis = 1).copy()
```

```python
Y = credit_downsampled['default'].copy()
```

# **One-Hot Encoding**

```python
# using get_dummies
# columns_to_encode = ['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
x_encoded = pd.get_dummies(X, columns = ['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6'], dtype='uint8')
```

```python
x_encoded
```

```python
x_encoded.dtypes
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_encoded, Y, test_size = 0.3, random_state = 18)
```

```python
#len(x_train)
#len(x_test)
len(y_train)
#len(y_test)
```

```python
from sklearn.preprocessing import scale

x_train_scaled = scale(x_train)
x_test_scaled = scale(x_test)
```

```python
from sklearn.svm import SVC
clf_svm = SVC(random_state = 81)
clf_svm.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
conf_matrix = ConfusionMatrixDisplay.from_estimator(clf_svm, x_test_scaled, y_test, display_labels= ['Did not Default', 'Default'])
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = [{
    'C' : [0.1, 1, 10, 100, 1000],
    'gamma': ['scale', 0.1, 0.01, 0.001, 0.0001],
    'kernel' : ['rbf']}]
optimal_params = GridSearchCV(clf_svm, param_grid, cv = 5, scoring = 'accuracy', verbose = 0)
optimal_params.fit(x_train_scaled, y_train)
print(optimal_params.best_params_)
```

```python
clf_svm_optimal = SVC(C = 1, gamma = 0.01, kernel = 'rbf', random_state = 77)
clf_svm_optimal.fit(x_train_scaled, y_train)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
conf_matrix = ConfusionMatrixDisplay.from_estimator(clf_svm_optimal, x_test_scaled, y_test, display_labels= ['Did not Default', 'Default'])
```
