```python
emails = pd.read_csv('email.csv')
emails.head()
```

```python
#data exploration
emails.duplicated().sum()
```

```python
emails.drop_duplicates(inplace = True)
```

```python
emails.isna().sum()
```

```python
emails.dtypes
```

```python
emails.dropna(inplace = True)
```

```python
# converting the category column to int datatype
emails['Spam'] = emails['Category'].map( {'spam' : 1, 'ham' : 0}).astype(int)
emails.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(emails['Message'], emails['Spam'], test_size = 0.3, random_state = 21)
```

```python
# we need to convert messages to numbers
# we can use vectorizer for that
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
x_train_vectorized = vectorizer.fit_transform(x_train.values)
print(vectorizer.get_feature_names_out())
#x_train_vectorized.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_vectorized, y_train)
```

```python
# testing on test data
# vectorizing the test data
x_test_vectorized = vectorizer.transform(x_test)
model.predict(x_test_vectorized)
```

```python
model.score(x_test_vectorized, y_test)
```

```python
some_emails = [
    'Heyy, stop wasting your time on watching movies',
    'John, we noticed you are not taking advantage of your Dropbox account.',
    'Hey Karen, we havent seen you in a while and need to know if you are still interested in seeing emails from us.',
    'Elevate your date nights& make your movie sessions more special with Samsung!'
]
```

```python
some_emails_vectorized = vectorizer.transform(some_emails)
model.predict(some_emails_vectorized)
```

# **Using Pipelines**

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('v', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
clf.predict(some_emails)
```
